```python
def check_answers(correct_answers, student_answers):
    total_questions = len(correct_answers)
    correct_count = 0
    incorrect_count = 0
    unanswered_count = 0

    for i in range(total_questions):
        if student_answers[i] == "":
            unanswered_count += 1
        elif student_answers[i] == correct_answers[i]:
            correct_count += 1
        else:
            incorrect_count += 1

    return correct_count, incorrect_count, unanswered_count

Questions = {
    "1": "What is part of a database that holds only one type of information?",
    "2": "'OS' computer abbreviation usually means?",
    "3": "What was the active medium used in the first working laser ever constructed?",
    "4": "'.MOV' extension refers usually to what kind of file?",
    "5": "What does EPROM stand for?"
}
correct_answers = []
for i in range(1, len(Questions) + 1):
    correct_answer = input(f"Enter the correct answer for Question {i}: ")
    correct_answers.append(correct_answer)

student_answers = []
for i in range(1, len(Questions) + 1):
    student_answer = input(f"Enter the student's answer for Question {i}: ")
    student_answers.append(student_answer)

correct_count, incorrect_count, unanswered_count = check_answers(correct_answers, student_answers)

print("\nResults:")
print("Correct Answers:", correct_count)
print("Incorrect Answers:", incorrect_count)
print("Unanswered Questions:", unanswered_count)

total_questions = len(correct_answers)
score = (correct_count / total_questions) * 100
print("Student's Score: {:.2f}%".format(score))
```

```python
class ShoppingAssistant:
    def __init__(self):
        self.available_products = {
            "Smartphone": 500,
            "Headphones": 100,
            "Laptop": 800,
            "Camera": 300
        }
        self.wallet_balance = 1000
        self.shopping_cart = []

    def get_user_preferences(self):
        print("Welcome to the Online AI Shopping Assistant!")
        print("Available products:")
        for product, price in self.available_products.items():
            print(f"{product}: {price} credits")

        while True:
            user_input = input("Enter the product you want to buy ")

            if user_input in self.available_products:
                self.shopping_cart.append(user_input)

    def recommend_products(self):
        print("Recommendations based on your preferences:")
        for product in self.available_products:
            print(product)

    def make_payment(self):
        total_amount = sum(self.available_products[product] for product in self.shopping_cart)
        if self.wallet_balance >= total_amount:
            self.wallet_balance -= total_amount
            print("Payment successful. Order placed!")
        else:
            print("Insufficient balance. Payment failed.")
        print(f"Remaining wallet balance: {self.wallet_balance} credits")
assistant = ShoppingAssistant()
assistant.get_user_preferences()
assistant.recommend_products()
assistant.make_payment()
```

```python
from difflib import SequenceMatcher

def calculate_similarity(text1, text2):
    similarity_ratio = SequenceMatcher(None, text1, text2).ratio()
    return similarity_ratio

database_texts = [
    "This is a sample text for comparison.",
    "Plagiarism is a serious academic offense.",
    "Online plagiarism checkers can be helpful.",
    "Avoiding plagiarism is important in academic writing."
]

def check_plagiarism(input_text):
    similarities = []
    for text in database_texts:
        similarity_ratio = calculate_similarity(input_text, text)
        similarities.append((similarity_ratio, text))

    return similarities

while True:
    user_input = input("Enter the assignment text to check for plagiarism (or 'q' to quit): ")

    if user_input.lower() == 'q':
        print("Exiting the plagiarism checker.")
        break

    try:

        similarity_results = check_plagiarism(user_input)
        
        print("\nSimilarity Results:")
        for similarity, text in similarity_results:
            print(f"Similarity ratio: {similarity:.2f}, Text: {text}")
        print()
    except Exception as e:
        print(f"Error: {str(e)}")
```

```python
import random

def generate_time_table():
    days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
    class_timings = ["9:00 AM - 9:45 AM", "10:00 AM - 10:45 AM", "11:00 AM - 11:45 AM", "12:00 PM - 12:45 PM", "1:00 PM - 1:45 PM"]
    subjects = ["Math", "English", "Science", "History", "Physical Education"]
    teachers = ["Dr.Madhusudanan", "Dr.Auxillia", "Raajan", "Pragash Kaliyan", "Lakshmipriya"]

    time_table = {}

    for day in days_of_week:
        time_table[day] = {}

        available_subjects = subjects.copy()
        available_teachers = teachers.copy()

        for timing in class_timings:
            subject = random.choice(available_subjects)
            teacher = random.choice(available_teachers)

            time_table[day][timing] = {"Subject": subject, "Teacher": teacher}

            available_subjects.remove(subject)
            available_teachers.remove(teacher)

    return time_table

def print_time_table(time_table):
    for day, timings in time_table.items():
        print(day)
        for timing, info in timings.items():
            print(f"{timing}: {info['Subject']} - {info['Teacher']}")
        print()

if __name__ == "__main__":
    time_table = generate_time_table()
    print_time_table(time_table)
```
